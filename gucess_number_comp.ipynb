```python
import random

EASY_LEVEL_ATTEMPT = 10
HARD_LEVEL_ATTEMPT = 5

def set_difficulty(level_chosen):
    if level_chosen == "easy":
        return EASY_LEVEL_ATTEMPT
    elif level_chosen == "hard":
        return HARD_LEVEL_ATTEMPT
    else:
        return None  

def check_answer(guess_number, answer, attempt, last_guess):
    if guess_number < answer:
        if last_guess is not None and guess_number < last_guess:
            print("You are going even lower! Try a higher number. üîº")
        else:
            print("Your guess is too low. Try a higher number! ‚¨ÜÔ∏è")
        return attempt - 1
    elif guess_number > answer:
        if last_guess is not None and guess_number > last_guess:
            print("You are going even higher! Try a lower number. üîΩ")
        else:
            print("Your guess is too high. Try a lower number! ‚¨áÔ∏è")
        return attempt - 1
    else:
        print(f"üéâ You guessed it right! The answer was {answer} üéØ")
        return -1  

def game():
    ascii_art = """
   ____                                              _     _                                            _
  / ___|  _   _    ___    ___    ___   ___   ___    | |_  | |__     ___     _ __    _   _   _ __ ___   | |__     ___   _ __
 | |  _  | | | |  / __|  / __|  / _ \ / __| / __|   | __| | '_ \   / _ \   | '_ \  | | | | | '_ ` _ \  | '_ \   / _ \ | '__|
 | |_| | | |_| | | (__  | (__  |  __/ \__ \ \__ \   | |_  | | | | |  __/   | | | | | |_| | | | | | | | | |_) | |  __/ | |
  \____|  \__,_|  \___|  \___|  \___| |___/ |___/    \__| |_| |_|  \___|   |_| |_|  \__,_| |_| |_| |_| |_.__/   \___| |_|
    """
    print(ascii_art)

    print("Let me think of a number between 1 to 50")
    answer = random.randint(1, 50)

    level = input("Choose difficulty level... Type 'easy' or 'hard': ").lower()
    attempt = set_difficulty(level)

    if attempt is None:
        print("Invalid difficulty level! Please restart the game.")
        return

    last_guess = None 

    while attempt > 0:
        print(f"You have {attempt} attempts remaining to guess the number")
        guess_number = int(input("Make a guess: "))

        attempt = check_answer(guess_number, answer, attempt, last_guess)
        last_guess = guess_number  

        if attempt == -1: 
            return 
        elif attempt == 0:  
            print(f"You are out of guesses.. You Lose! üò¢ The correct number was {answer}")
            return
        else:
            print("Guess Again")

game()
```
