```python
import random

word_list = ['apple','orange','grapes','banana']

lives = 6

choicen_word = random.choice(word_list)
print(choicen_word)

display = []

for i in range(len(choicen_word)):
    display += '_'
print(display)

game_over = False
while not game_over: 
    guess_letter = input('Guess a letter: ').lower()
    for position in range(len(choicen_word)):
        letter = choicen_word[position]
        if letter == guess_letter:
            display[position] = guess_letter
    print(display)  
    if guess_letter not in choicen_word:
        lives -= 1
        print(f"Wrong guess! Remaining lives: {lives}")
        if lives == 0:
            game_over = True
            print(f"Game over")
    if "_" not in display:
        game_over = True
        print("You win!")
```
